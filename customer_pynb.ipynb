```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, LabelEncoder
data = {
    'CustomerID': [101, 102, 103, 104, 105, 106, 107],
    'Name': ['John', 'Mary', 'Alex', 'Sara', 'John', 'Emma', 'Liam'],
    'Age': [28, 32, 40, 25, 28, 35, 29],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'New York', 'Phoenix', 'Philadelphia'],
    'PurchaseAmount': [250, 450, 300, 150, 250, 500, 200]
}
df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)
duplicates = df.duplicated()
print("\nDuplicate rows:")
print(df[duplicates])
df_clean = df.drop_duplicates()
print("\nDataFrame after removing duplicates:")
print(df_clean)
print("\nSummary statistics:")
print(df_clean.describe())
plt.figure(figsize=(8,5))
sns.barplot(x='Name', y='PurchaseAmount', data=df_clean)
plt.title('Purchase Amount by Customer')
plt.show()
label_encoders = {}
for col in ['Name', 'City']:
    le = LabelEncoder()
    df_clean[col] = le.fit_transform(df_clean[col])
    label_encoders[col] = le
print("\nDataFrame after encoding categorical variables:")
print(df_clean)
scaler = StandardScaler()
df_clean[['Age', 'PurchaseAmount']] = scaler.fit_transform(df_clean[['Age', 'PurchaseAmount']])
print("\nDataFrame after scaling numerical features:")
print(df_clean)
```
